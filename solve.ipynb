```kotlin
val Ik1 = 0.91
```

```kotlin

fun computeB(I: Double): Double{
    val N = 2006.0
    val L = 167.0 * 10.0.pow(-3)
    val D = 62.0 * 10.0.pow(-3)
    val nu = 4.0 * PI * 10.0.pow(-7)

    return (nu * N * I) / (sqrt(L.pow(2) + D.pow(2)))
}
```

```kotlin
val b1 = computeB(Ik1)
b1
```

```kotlin
fun computeEfracM(U: Double, B: Double, ): Double {
    val ra = 6.0 * 10.0.pow(-3) // радиус анода
    return (8.0 * U)/(B * ra).pow(2)

}
```

```kotlin
val em1 = computeEfracM(60.0, b1)
em1
```

```kotlin
val Ik2 = 1.025
val b2 = computeB(Ik2)
b2
```

```kotlin
val em2 = computeEfracM(90.0, b2)
em2
```
